```python
bat_rows_list = []
for row in bat_rows[1:]:
    cur_row = []
    for c in row:
        try:
            cur_row.append(c.text)
        except:
            cur_row.append('')
    if len(cur_row) == 30:
        bat_rows_list.append(cur_row)
```

```python
batters = pd.DataFrame(bat_rows_list, columns=b_header)
```

```python
c = batters.columns.tolist()
c[-1] = 'pos_summary'
c = [x.lower() for x in c]
batters.columns = c
non_num = ['name','tm','lg','pos_summary']
num_col = [x for x in c if x not in non_num]
for col in num_col:
    batters[col] = pd.to_numeric(batters[col])
```
