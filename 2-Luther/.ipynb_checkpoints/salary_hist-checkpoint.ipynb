```python
from train_test_fns import *

import matplotlib as mpl
mpl.rcParams['figure.dpi']= 300
```

```python
a,b,c,d = load_and_split_data()
a.shape
```

```python
c.shape
```

```python
full_train = a.join(b)
full_test = c.join(d)
```

```python
full_df = pd.concat([full_train,full_test])
```

```python
full_df.shape
```

```python
full_df['salary'] = full_df['log10_adj'].apply(lambda x: 10**x)
```

```python
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(4,5))
ax1.hist(full_df['salary']/1000000,bins=15,color='g')
ax2.hist(full_df['log10_adj'],bins=15,color='b')
#fig.suptitle("Histogram of Salaries")
ax1.set_ylabel('Counts')
ax2.set_ylabel('Counts')
ax1.set_xlabel("Salary (Millions)")
ax2.set_xlabel('log(Salary)');
ax1.set_yticks([], [])
ax2.set_yticks([], [])

plt.tight_layout()
plt.savefig('salaryhist.png')
```
