```python
import pandas as pd
import numpy as np


filename = 'subway_data.txt'
s = pd.read_csv(filename)
```

```python
s.columns = s.columns.str.strip()
```

```python
s['datetime'] = s[['DATE','TIME']].apply(lambda x: ' '.join(x), axis=1)
s['datetime'] = pd.to_datetime(s['datetime'])
```

```python
s.drop(['DATE','TIME','DIVISION'],axis=1,inplace=True)
```
